```python
# Start handshake/heartbeat/control stream
from drone import start_drone
import time

drone = start_drone(verbose=True)
```

```python
# Example: small throttle bump
drone.takeoff()
drone.set_axes(x=0x70)
time.sleep(1.0)
drone.set_axes(z=0x90)
time.sleep(1.0)
drone.neutral()
time.sleep(2.0)
drone.set_axes(z=0x70)
time.sleep(1.0)
drone.land()
```

```python
drone.estop()
```
